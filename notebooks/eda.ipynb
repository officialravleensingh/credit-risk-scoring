# Credit Risk Scoring - Exploratory Data Analysis

## Objective
Analyze the credit risk dataset to understand patterns, distributions, and relationships between features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Data

```python
df = pd.read_csv('../dataset/original_dataset.csv')
print(f"Dataset shape: {df.shape}")
df.head()
```

## 2. Basic Information

```python
df.info()
```

```python
df.describe()
```

## 3. Missing Values

```python
missing = df.isnull().sum()
print(f"Total missing values: {missing.sum()}")
print(missing[missing > 0])
```

## 4. Target Variable Analysis

```python
target_counts = df['loan_paid_back'].value_counts()
print(target_counts)
print(f"\nPaid back rate: {(target_counts[1] / len(df)) * 100:.2f}%")
print(f"Default rate: {(target_counts[0] / len(df)) * 100:.2f}%")
```

```python
plt.figure(figsize=(8, 6))
colors = ['#ff6b6b', '#51cf66']
labels = ['Defaulted', 'Paid Back']
plt.pie(target_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Loan Repayment Distribution', fontsize=14, fontweight='bold')
plt.show()
```

## 5. Numerical Features Distribution

```python
num_features = ['age', 'annual_income', 'credit_score', 'loan_amount', 'debt_to_income_ratio']

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, col in enumerate(num_features):
    axes[idx].hist(df[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    axes[idx].set_title(f'{col} Distribution', fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Frequency')

axes[5].axis('off')
plt.tight_layout()
plt.show()
```

## 6. Credit Score vs Loan Repayment

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='credit_score', by='loan_paid_back', ax=plt.gca())
plt.title('Credit Score vs Loan Repayment')
plt.suptitle('')
plt.xlabel('Loan Paid Back (0=Default, 1=Paid)')
plt.ylabel('Credit Score')
plt.show()
```

## 7. Income vs Loan Amount

```python
plt.figure(figsize=(10, 6))
colors_map = {0: 'red', 1: 'green'}
for status in [0, 1]:
    subset = df[df['loan_paid_back'] == status]
    plt.scatter(subset['annual_income'], subset['loan_amount'], 
               alpha=0.5, label=f"{'Defaulted' if status == 0 else 'Paid Back'}")
plt.xlabel('Annual Income')
plt.ylabel('Loan Amount')
plt.title('Income vs Loan Amount by Repayment Status')
plt.legend()
plt.show()
```

## 8. Categorical Features Analysis

```python
cat_features = ['gender', 'marital_status', 'education_level', 'employment_status']

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

for idx, col in enumerate(cat_features):
    df[col].value_counts().plot(kind='bar', ax=axes[idx], color='coral')
    axes[idx].set_title(f'{col} Distribution', fontweight='bold')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('Count')
    axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 9. Employment Status vs Loan Repayment

```python
employment_loan = pd.crosstab(df['employment_status'], df['loan_paid_back'], normalize='index') * 100
employment_loan.plot(kind='bar', figsize=(10, 6), color=['#ff6b6b', '#51cf66'])
plt.title('Loan Repayment Rate by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Percentage')
plt.legend(['Defaulted', 'Paid Back'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 10. Correlation Analysis

```python
num_cols = ['age', 'annual_income', 'monthly_income', 'debt_to_income_ratio', 
            'credit_score', 'loan_amount', 'interest_rate', 'loan_term', 
            'num_of_open_accounts', 'total_credit_limit', 'current_balance', 
            'delinquency_history', 'public_records', 'num_of_delinquencies', 'loan_paid_back']

correlation = df[num_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, fmt='.2f', 
            square=True, linewidths=1)
plt.title('Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 11. Key Insights

### Findings:
1. **Class Distribution**: 79.99% loans paid back, 20.01% defaulted (imbalanced)
2. **Credit Score**: Strong predictor - higher scores correlate with repayment
3. **No Missing Values**: Clean dataset, no imputation needed
4. **Income Distribution**: Right-skewed, some high earners
5. **Debt-to-Income Ratio**: Important factor in default prediction
6. **Employment Status**: Employed borrowers have higher repayment rates
7. **Delinquency History**: Strong negative correlation with repayment

### Recommendations:
- Credit score is the most important feature
- Delinquency history and public records are strong indicators
- Consider class imbalance in model training
- Feature scaling needed for numerical features
- Label encoding required for categorical features
